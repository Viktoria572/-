Завдання 2.1 та 2.2 Загрузити датасет, що містить інформацію про відвідуваність музеїв в різні роки. Вивести перші 5 рядків датасету.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантаження датасету з Google Colab
file_path = ('/content/sample_data/museum_visitors.csv')

# Завантаження даних
df = pd.read_csv(file_path)

# Виведення перших 5 рядків датасету
df.head()
```

Завдання 2.3. Провести попередній аналіз даних (визначити розмір датасету, тип даних, кількість пропусків).

```python
# Визначення розміру датасету (кількість рядків та стовпців)
shape = df.shape

# Визначення типів даних в кожному стовпці
data_types = df.dtypes

# Підрахунок кількості пропусків в кожному стовпці
missing_values = df.isnull().sum()

# Виведення результатів аналізу
print("Розмір датасету:", shape)
print("Типи даних:")
print(data_types)
print("Кількість пропусків:")
print(missing_values)
```

Завдання 2.4. Перевірити наявність дублікатів. Якщо є, видалити їх і вивести розмір таблиці (без дублікатів).

```python
# Перевірка наявності дублікатів
initial_shape = df.shape  # Розмір початкового датасету
df_no_duplicates = df.drop_duplicates()  # Видалення дублікатів

# Визначення розміру таблиці після видалення дублікатів
final_shape = df_no_duplicates.shape

# Виведення результатів
print("Розмір початкової таблиці:", initial_shape)
print("Розмір таблиці без дублікатів:", final_shape)
```

Завдання 2.5. Про які роки містить інформацію датасет.

```python
# Визначення унікальних років у датасеті
unique_years = df['Date'].str.extract(r'(\d{4})')[0].unique()

# Виведення унікальних років
print("Унікальні роки у датасеті:")
for year in unique_years:
    print(year)
```

Завдання 2.6. Змінити назви столбців. Замінити великі букви на маленьки, пробіли на підкреслення (date, avila_adobe, firehouse_museum, chinese_american_museum, america_tropical_interpretive_center).

```python
# Зміна назв стовпців на малі букви та заміна пробілів на підкреслення
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Виведення оновлених назв стовпців
print("Оновлені назви стовпців:")
print(df.columns)
```

Завдання 2.7. Обчислити середню кількість відвідувачів для кожного музею протягом всього періоду.

```python
# Обчислення середньої кількості відвідувачів для кожного музею
average_visitors = df.mean()

# Виведення результатів
print("Середня кількість відвідувачів для кожного музею:")
print(average_visitors)
```

Завдання 2.8. Знайти мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік

```python
# Вибір даних для 2018 року
data_2018 = df[df['year'] == 2018]

# Знаходження мінімальної та максимальної кількості відвідувачів для кожного музею у 2018 році
min_visitors_2018 = data_2018[['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum', 'America Tropical Interpretive Center']].min()
max_visitors_2018 = data_2018[['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum', 'America Tropical Interpretive Center']].max()

# Виведення результатів
print("Мінімальна кількість відвідувачів для кожного музею у 2018 році:")
print(min_visitors_2018)

print("\nМаксимальна кількість відвідувачів для кожного музею у 2018 році:")
print(max_visitors_2018)
```

Завдання 2.9. Визначити місяці з найвищою і найнижчою загальною кількістю відвідувачів серед усіх музеїв для 2015 року. Вивести назви місяців (не цифри).

```python
# Перетворення стовпця 'Date' у формат дати
data['Date'] = pd.to_datetime(data['Date'])

# Вибір даних за 2015 рік
data_2015 = data[data['Date'].dt.year == 2015]

# Групування за місяцями і обчислення загальної кількості відвідувачів
monthly_visitors = data_2015.groupby(data_2015['Date'].dt.month)['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum', 'America Tropical Interpretive Center'].sum().sum(axis=1)

# Знаходимо місяць з найвищою і найнижчою кількістю відвідувачів
highest_month = monthly_visitors.idxmax()
lowest_month = monthly_visitors.idxmin()

# Вивід назв місяців (не цифри)
month_names = ['січень', 'лютий', 'березень', 'квітень', 'травень', 'червень', 'липень', 'серпень', 'вересень', 'жовтень', 'листопад', 'грудень']

highest_month_name = month_names[highest_month - 1]  # Відніміть 1, оскільки індекси починаються з 0
lowest_month_name = month_names[lowest_month - 1]

print(f"Місяць з найвищою кількістю відвідувачів у 2015 році: {highest_month_name}")
print(f"Місяць з найнижчою кількістю відвідувачів у 2015 році: {lowest_month_name}")
```

Завдання 2.10. Порівняти кількість відвідувачів музею "Avila Adobe" у літні і зимові місяці 2018 року

```python
# Визначення літніх і зимових місяців
summer_months = [6, 7, 8]  # Червень, липень, серпень
winter_months = [12, 1, 2]  # Грудень, січень, лютий

# Фільтрація даних для 2018 року і музею "Avila Adobe"
avila_adobe_2018 = df[(df['year'] == 2018) & (df['Avila Adobe'] == 'Avila Adobe')]

# Обчислення кількості відвідувачів у літні та зимові місяці
summer_visitors = avila_adobe_2018[avila_adobe_2018['year'].isin(summer_months)]['Avila Adobe'].sum()
winter_visitors = avila_adobe_2018[avila_adobe_2018['year'].isin(winter_months)]['Avila Adobe'].sum()

# Виведення результатів
print("Кількість відвідувачів музею 'Avila Adobe' в літні місяці 2018 року:", summer_visitors)
print("Кількість відвідувачів музею 'Avila Adobe' в зимові місяці 2018 року:", winter_visitors)
```

Завдання 2.11. Знайти кореляцію між кількістю відвідувачів в кожному з музеїв  та датами у 2016 році

Завдання 2.12. Візуалізація даних
Побудувати теплокарту кореляції для попереднього завдання.

```python
# Фільтрація даних для 2016 року
data_2016 = df[df['year'] == 2016]

# Обчислення кореляції між кількістю відвідувачів у різних музеях та датами
correlation = data_2016.corr()

# Побудова теплокарти кореляції
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Теплокарта кореляції')
plt.show()
```

Завдання 2.13. Побудувати графік ліній для кожного музею, показуючи їхню відвідуваність протягом 2017 року.

```python
# Фільтрація даних для 2017 року
data_2017 = df[df['year'] == 2017]

# Визначення стовпців, які представляють музеї (замість цих назв вставте правильні назви стовпців)
museum_columns = ['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum', 'America Tropical Interpretive Center']

# Побудова графіка ліній для кожного музею
plt.figure(figsize=(12, 6))
for museum in museum_columns:
    plt.plot(data_2017['year'], data_2017[museum], label=museum)

plt.xlabel('Рік')
plt.ylabel('Кількість відвідувачів')
plt.title('Відвідуваність музеїв у 2017 році')
plt.legend(loc='upper right')
plt.grid(True)
plt.xticks(rotation=45)  # Поворот підписів по осі X для зручності
plt.show()
```

Завдання 2.14. Побудувати графіки розсіювання для кожного музею за 2018 рік

```python
# Перетворення стовпця 'Date' у формат дати
data['Date'] = pd.to_datetime(data['Date'])

# Вибір даних за 2018 рік
data_2018 = data[data['Date'].dt.year == 2018]

# Список музеїв
museums = ['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum', 'America Tropical Interpretive Center']

# Побудова графіків розсіювання для кожного музею
plt.figure(figsize=(12, 8))

for museum in museums:
    plt.scatter(data_2018['Date'], data_2018[museum], label=museum, alpha=0.5)

plt.title('Графіки розсіювання для музеїв у 2018 році')
plt.xlabel('Дата')
plt.ylabel('Кількість відвідувачів')
plt.legend()
plt.grid(True)
plt.show()
```

Завдання 2.15. Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий.

```python
# Перетворення стовпця 'Date' у формат дати
data['Date'] = pd.to_datetime(data['Date'])

# Розділімо дані за роками
data_2014 = data[data['Date'].dt.year == 2014]
data_2015 = data[data['Date'].dt.year == 2015]
data_2016 = data[data['Date'].dt.year == 2016]
data_2017 = data[data['Date'].dt.year == 2017]

# Визначення кольорів для графіків
colors = ['blue', 'green', 'yellow', 'gray']

# Побудова гістограм
plt.figure(figsize=(12, 8))
for i, dataset in enumerate([data_2014, data_2015, data_2016, data_2017]):
    plt.subplot(2, 2, i+1)
    plt.hist(dataset['Avila Adobe'], color=colors[i])
    plt.title(f'Рік {2014 + i}')
    plt.xlabel('Кількість відвідувачів')
    plt.ylabel('Кількість місяців')

plt.tight_layout()
plt.show()
```
